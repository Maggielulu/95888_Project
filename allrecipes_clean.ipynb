```python
import os
import sys
import bs4
import requests
import time
import re
import csv
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
#read the raw file and revert it back to a dictionary
raw_csv = pd.read_csv("allrecipes_raw.csv", index_col=0)
raw_dict = raw_csv.to_dict("split")
raw_dict = dict(zip(raw_dict["index"], raw_dict["data"]))

id = 0
ls = ([["entry_id","drink_name","ingredient","measurement"]])
for key in raw_dict:
    vl = raw_dict.get(key)
    whitelist = set('abcdefghijklmnopqrstuvwxy ABCDEFGHIJKLMNOPQRSTUVWXYZ')
    key = ''.join(filter(whitelist.__contains__, key))
    for v in vl:
        if pd.isnull(v) == False:
            #regular expression to deal with number/fractions + list of usual measurements
            n = re.search(r'[\d/\.]+(.*((cup|tablespoon|can|ounce|teaspoon|pint|jigger|shot)s?|(dash|glass)(es)?))?', v)
            if pd.isnull(n):
                ls.append([id,key,v,'N/A'])
            else:
                #split the measurement from the ingredients, in fact I am thinking perhaps I 
                #can further separate the units and numbers too so that the clustering can be more accurate?
                #add it to the to-do list
                ls.append([id,key,"".join(v.rsplit(n.group())),n.group()])
            id = id + 1;
            
            
#output the cleaned csv
with open("allrecipes_clean.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerows(ls)
```
